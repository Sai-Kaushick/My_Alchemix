```python
df=pd.read_csv('Training data.csv',sep=';')
df.head()
```

```python
new_df = df[['Smiles', 'Standard Value','Standard Type']]
new_df=new_df.dropna()
```

```python
new_df
```

```python
def norm_value(input):
  norm = []
  for i in input['Standard Value']:
    if i > 100000000:
      i = 100000000
    norm.append(i)
  input['standard_value_norm'] = norm
  x = input.drop(columns=['Standard Value'])
  return(x)
standard_df = norm_value(new_df)
standard_df
```

```python
if 'standard_value_norm' in new_df:
  new_df['standard_value_M'] = new_df['standard_value_norm'] * 1e-9
  new_df.to_csv('Training data.csv', index=False)
if 'standard_value_norm' in new_df:
    new_df['pIC50'] = np.where(new_df['standard_value_M'] > 0,-np.log10(new_df['standard_value_M']),np.nan)
    new_df.to_csv('path_to_your_updated_data.csv', index=False)

    print(new_df.head())
```

```python

Pic50_threshold=[]

for i in new_df.pIC50:
  if float(i) >= 7:
      Pic50_threshold.append("Rank I")
  elif float(i) >=5 :
      Pic50_threshold.append("Rank II")
  else:
      Pic50_threshold.append("Rank III")
```
